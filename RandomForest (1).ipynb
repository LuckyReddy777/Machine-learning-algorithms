```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('horse.csv')
```

```python
data.head()
```

```python
data['outcome'].unique()
```

```python
categorical_columns = data.columns[data.dtypes=='object']
```

```python
categorical_columns
```

```python
features = data.drop(['outcome'], axis = 1)
target = data['outcome']
```

```python
features = pd.get_dummies(features)
```

```python
features
```

```python
features.columns
```

```python
print(features.isna().sum())
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(missing_values=np.NaN, strategy='most_frequent')
```

```python
features = imp.fit_transform(features)
```

```python
pd.DataFrame(features).isna().any().any()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
lbl_encoder=LabelEncoder()
target = lbl_encoder.fit_transform(target)
```

```python
X_train,X_test,y_train,y_test = train_test_split(features,target)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
my_dt_classifier = DecisionTreeClassifier()
```

```python
my_dt_classifier.fit(X_train,y_train)
```

```python
my_dt_predictions = my_dt_classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(my_dt_predictions,y_test)
```

```python
confusion_matrix(my_dt_predictions,y_test)
```

```python
print(classification_report(my_dt_predictions,y_test))
```
