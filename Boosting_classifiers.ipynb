```python
import pandas as pd
```

```python
data = pd.read_csv('pima-indians-diabetes.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.dtypes
```

```python
features = data.drop(['Outcome'], axis = 1)
target = data[['Outcome']]
features.shape, target.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state = 101, train_size = 0.8)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
my_dt_model = DecisionTreeClassifier()
```

```python
my_dt_model.fit(X_train,y_train)
my_dt_preds = my_dt_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print(accuracy_score(y_test, my_dt_preds))
```

```python
print(confusion_matrix(y_test, my_dt_preds))
```

```python
from sklearn.ensemble import RandomForestClassifier

my_rf_classifier = RandomForestClassifier()

my_rf_classifier.fit(X_train,y_train)
my_rf_preds = my_rf_classifier.predict(X_test)
```

```python
print(accuracy_score(y_test, my_rf_preds))
```

```python
print(confusion_matrix(y_test, my_rf_preds))
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
my_ada_clf = AdaBoostClassifier(n_estimators=50)
```

```python
my_ada_clf.fit(X_train,y_train)
my_ada_preds = my_ada_clf.predict(X_test)
```

```python
print(accuracy_score(y_test, my_ada_preds))
```

```python
print(confusion_matrix(y_test, my_ada_preds))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
my_gb_clf = GradientBoostingClassifier()
```

```python
my_gb_clf.fit(X_train, y_train)
```

```python
my_gb_preds = my_gb_clf.predict(X_test)
```

```python
print(accuracy_score(y_test, my_gb_preds))
```

```python
from xgboost import XGBClassifier, XGBRFClassifier
```

```python
my_xgb_clf = XGBRFClassifier(booster = 'gbtree')
```

```python
my_xgb_clf.fit(X_train, y_train)
```

```python
my_xgb_preds = my_xgb_clf.predict(X_test)
```
