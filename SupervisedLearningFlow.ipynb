```python
import pandas as pd
from sklearn.datasets import load_iris
```

## Data Acquisition

```python
iris_data = load_iris()
```

```python
input_data = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
input_data
```

```python
iris_data.target_names
```

```python
output_data = pd.DataFrame(iris_data.target, columns = ['Target'])
output_data
```

```python
print(input_data.shape, output_data.shape)
```

## Splitting the Data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(input_data, output_data, random_state = 30)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Training your model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
my_model = DecisionTreeClassifier()
```

```python
# Takes the input and output of training i.e. X_train and y_train
my_model.fit(X_train, y_train)
```

#### Validation Part

```python
y_pred = my_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```
