```python
import pandas as pd
```

```python
data = pd.read_csv('Tweets.csv')
data.head()
```

```python
data = data[['text','airline_sentiment']]
data.head()
```

```python
import re, itertools
```

```python
# Writing regular expression to extract all tags starting with @
r = re.compile(r"([@])(\w+)\b")
```

```python
# Lambda function to run through a file with column as "text"
references = map(lambda x : r.findall(x), data['text'])
references
```

```python
# Extraction of tags
all_references = set(list(itertools.chain.from_iterable(references)))
all_references
```

```python
# Combining @ with the tags
References = map(lambda x:x[0]+x[1],all_references)
References
```

```python
# Writing to a file
file = open('References.txt','a')
for each in References:
    file.write(each+'\n')
file.close()
```
